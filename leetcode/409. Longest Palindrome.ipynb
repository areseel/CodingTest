```python
class Solution:
    def longestPalindrome(self, s: str) -> int:
        counts = Counter(s)
        output = 0
        center = False

        for key in counts:
            if counts[key] % 2 == 0:
                output += counts[key]
            elif counts[key] % 2 == 1:
                output += (counts[key] - 1)
                if center == False:
                    output += 1
                    center = True

        return output
```
