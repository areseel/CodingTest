```python
class Solution:
    def reorganizeString(self, s: str) -> str:
        counts = collections.Counter(s)
        sorted_items = counts.most_common()

        if sorted_items[0][1] > (len(s)+1) // 2: return ''
        ans = [''] * len(s)
        i = 0

        for elem in range(len(sorted_items)):
            for _ in range(sorted_items[elem][1]):
                ans[i] = sorted_items[elem][0]
                i += 2
                if i >= len(s):
                    i = 1

        return ''.join(ans)
```
